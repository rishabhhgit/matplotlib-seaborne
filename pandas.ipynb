```python
import pandas as pd
import numpy as np
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
columns=['name','age','city','salary']
y_1=pd.DataFrame(data_list,columns=columns)
print(y_1)
y_1.drop('city',axis=1,inplace=True)
```
